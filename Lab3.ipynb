```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("USA_Housing.csv")
```

```python
dataset.mean(numeric_only=True)
```

```python
dataset.median(numeric_only=True)
```

```python
dataset.std(numeric_only=True)
```

```python
dataset.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(dataset.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
numeric_dataset = dataset.select_dtypes(include=np.number)
```

```python
sns.scatterplot(x=numeric_dataset.iloc[:, 0], y=numeric_dataset.iloc[:, 1])
plt.show()
```

```python
sns.boxplot(data=numeric_dataset)
plt.show()
```

```python
sns.violinplot(data=numeric_dataset)
plt.show()
```
